##Numerical Analysis

```python
# Numerical analysis and exploration
numerical_cols = ['Fee Structure','Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

##HeatMap

```python
# Select numerical columns for correlation analysis
numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending', 'Properties Owned']

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

# Exploratory Data Analysis (EDA) Report
ðŸ”¹ 1. Income and Financial Product Usage
Estimated Income shows a moderate positive correlation with most financial variables such as:

<li>Business Lending (0.33)

<li>Bank Loans (0.33)

<li>Credit Card Balance (0.30)

<li>Foreign Currency Accounts (0.31)

This suggests that higher-income individuals are more likely to:

<li>Engage in business activities

<li>Own more credit instruments

<li>Maintain international or diversified financial accounts

Implication: Income can be a useful feature when profiling financially active or high-value customers.

ðŸ”¹ 2. Strong Inter-Account Relationships
There are very strong correlations between:

<li>Bank Deposits and Checking Accounts (0.84)

<li>Bank Deposits and Saving Accounts (0.75)

<li>Checking Accounts and Saving Accounts (0.46)

These indicate that customers with one type of account tend to have others as well, reflecting financial discipline or wealth accumulation behaviors.

Implication: These features could be combined into a "Total Liquid Assets" metric to reduce redundancy and better capture financial liquidity.

ðŸ”¹ 3. Loan Behavior Patterns
Bank Loans are moderately correlated with:

<li>Business Lending (0.42)

<li>Credit Card Balance (0.37)

<li>Bank Deposits (0.36)

This shows that customers using loan services often:

<li>Participate in business-related borrowing

<li>Carry credit card balances

<li>Maintain deposit accounts, possibly to manage or secure loans

Implication: These customers may benefit from tailored credit management services or debt consolidation offers.

ðŸ”¹ 4. Age is Statistically Neutral
Age has virtually no correlation with any other numerical features in the dataset.

Implication: Age alone is not a driving factor in financial behavior within this dataset. It may be:

<li>Uniformly distributed

<li>Overshadowed by stronger predictors like income or account balances

<li>A non-linear influence, which basic correlation cannot detect

ðŸ”¹ 5. Property Ownership is an Isolated Variable
Properties Owned shows very weak or no linear correlation with any financial attributes.

Implication: This feature may:

<li>Be categorical (e.g., 0 or 1)

<li>Require segmented analysis

<li>Still hold predictive power in classification models despite its independence in correlation analysis

ðŸ”¹ 6. Risk of Multicollinearity
Some features, such as Bank Deposits, Checking Accounts, and Saving Accounts, are highly correlated with each other.

Implication:

<li>This could lead to multicollinearity problems in regression models.

<li>Consider removing redundant features or applying PCA to reduce dimensionality.
